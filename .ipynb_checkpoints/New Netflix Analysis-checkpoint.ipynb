## Exploratory Analysis and Visualization

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '00000000'
```

```python
xplt = df.groupby('country').count().sort_values('title', ascending=False).reset_index().country.head(10)
yplt = df.groupby('country').count().sort_values('title', ascending=False).reset_index().title.head(10)
plt.xticks(rotation=75)
plt.title(label='Number of Titles Available by Country')
sns.lineplot(x=xplt, y=yplt);
```

### We can plot how ratings have changed over time

```python
movie_ratings = df[df.type == 'Movie'].groupby('year_added')['averageRating'].mean()
```

```python
tv_ratings = df[df.type == 'TV Show'].groupby('year_added')['averageRating'].mean()
```

```python
sns.lineplot(x=movie_ratings.index, y=movie_ratings);
sns.lineplot(x=tv_ratings.index, y=tv_ratings);
plt.title(label='Movie and TV Show Ratings Over Time')
plt.legend(labels=['Movies', 'TV Shows']);
```

### TV shows appear to be performing much more strongly than movies in recent years.  Let's see if this observation holds across regions.  We can use IpyWidgets to make choosing a region easier.

```python
country_list = df[(df.country.isna() == False) & (df.country_count == 1)].country.sort_values().unique().tolist()

country_select = widgets.Dropdown(
                    options = country_list,
                    value = 'Argentina',
                    description = 'Countries',
                    disabled = False)

display(country_select)

button = widgets.Button(description="Plot")
output = widgets.Output()

display(button, output)

def on_button_clicked(b):
    avg_ratings_df = pd.DataFrame()
    country = country_select.value
    tv_avg = df[(df.country.str.contains(country)) & (df.type == 'TV Show')].groupby('year_added').averageRating.mean()
    movie_avg = df[(df.country.str.contains(country)) & (df.type == 'Movie')].groupby('year_added').averageRating.mean()
    with output:
        sns.lineplot(x=movie_avg.index, y=movie_avg);
        sns.lineplot(x=tv_avg.index, y=tv_avg);
        plt.title(label=country + ' Ratings Over Time')
        plt.legend(labels=['Movies', 'TV Shows']);

button.on_click(on_button_clicked)
```

### Has this trend affected the type of content Netflix has been adding?

```python
movies_added = df[df.type == 'Movie'].groupby('year_added').title.count()
tv_added = df[df.type == 'TV Show'].groupby('year_added').title.count()
sns.lineplot(x=movies_added.index, y=movies_added);
sns.lineplot(x=tv_added.index, y=tv_added);
plt.title(label='Type of Content Added by Netflix Over Time')
plt.legend(labels=['Movies', 'TV Shows']);
```

```python
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
shows = []
movies = []
for year in years:
    shows.append([year, df[(df.type == 'TV Show') & (df.year_added == year)].title.count()])
    movies.append([year, df[(df.type == 'Movie') & (df.year_added == year)].title.count()])
shows = np.array(shows)
movies = np.array(movies)
```

```python
# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = movies[:,1]
# Specify the values of orange bars (height)
orange_bar = shows[:,1]

# Position of bars on x-axis
ind = np.arange(2008,2021)

# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, blue_bar , width, label='Movies')
plt.bar(ind + width, orange_bar, width, label='TV Shows')

plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.title('Amount of Content Added by Type and Year')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, years)

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```

```python
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
shows = []
movies = []
for year in years:
    shows.append([year, df[(df.type == 'TV Show') & (df.year_added <= year)].title.count()])
    movies.append([year, df[(df.type == 'Movie') & (df.year_added <= year)].title.count()])
shows = np.array(shows)
movies = np.array(movies)
```

```python
sns.lineplot(x=movies[:,0], y=movies[:,1]);
sns.lineplot(x=shows[:,0], y=shows[:,1]);
plt.title(label='Amount of Content by Type and Year')
plt.legend(labels=['Movies', 'TV Shows']);
```

```python

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = movies[:,1]
# Specify the values of orange bars (height)
orange_bar = shows[:,1]

# Position of bars on x-axis
ind = np.arange(2008,2021)

# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.3       

# Plotting
plt.bar(ind, blue_bar , width, label='Movies')
plt.bar(ind + width, orange_bar, width, label='TV Shows')

plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.title('Amount of Content by Type and Year (Cumulative)')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, years)

# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```

## What is the distribution of titles?

### Answer:

```python
movie_countries = dict.fromkeys(df[(df.country_count == 1) & (df.country.isna() == False)].country.sort_values().unique().tolist())

for country in movie_countries.keys():
    movie_countries[country] = df[(df.country.str.contains(country)) & (df.type == 'Movie')].title.count()

movie_dist = pd.DataFrame.from_dict(movie_countries, orient='index', columns=['titles'])

plt.rc('font', size=12)
plt.figure(figsize=(20,15))
plt.title(label='Distribution of Movie Titles by Country', size=20)
sns.barplot(x=movie_dist.sort_values('titles', ascending=False).titles, y=movie_dist.sort_values('titles', ascending=False).index);
```

```python
tv_countries = dict.fromkeys(df[(df.country_count == 1) & (df.country.isna() == False)].country.sort_values().unique())

for country in tv_countries.keys():
    tv_countries[country] = df[(df.country.str.contains(country)) & (df.type == 'TV Show')].title.count()
    
tv_dist = pd.DataFrame.from_dict(tv_countries, orient='index', columns=['titles'])

plt.rc('font', size=12)
plt.figure(figsize=(20,15))
plt.title(label='Distribution of TV Shows by Country', size=20)
sns.barplot(x=tv_dist.sort_values('titles', ascending=False).titles, y=tv_dist.sort_values('titles', ascending=False).index);
```

## How do ratings vary by region?

### Answer:

```python
movie_ratings = dict.fromkeys(df[(df.country_count == 1) & (df.country.isna() == False)].country.sort_values().unique())

for country in movie_ratings.keys():
    movie_ratings[country] = df[(df.country.str.contains(country)) & (df.type == 'Movie')].averageRating.mean()

movie_ratings = pd.DataFrame.from_dict(movie_ratings, orient='index', columns=['ratings'])

plt.rc('font', size=12)
plt.figure(figsize=(20,15))
plt.title(label='Distribution of Average Movie Ratings by Country', size=20)
sns.barplot(x=movie_ratings.ratings.sort_values(ascending=False), y=movie_ratings.sort_values('ratings', ascending=False).index);
```

```python
tv_ratings = dict.fromkeys(df[(df.country_count == 1) & (df.country.isna() == False)].country.sort_values().unique())

for country in tv_ratings.keys():
    tv_ratings[country] = df[(df.country.str.contains(country)) & (df.type == 'TV Show')].averageRating.mean()

tv_ratings = pd.DataFrame.from_dict(tv_ratings, orient='index', columns=['ratings'])

plt.rc('font', size=12)
plt.figure(figsize=(20,15))
plt.title(label='Distribution of Average TV Show Ratings by Country', size=20)
sns.barplot(x=tv_ratings.ratings.sort_values(ascending=False), y=tv_ratings.sort_values('ratings', ascending=False).index);
```

## What are the most popular shows from each region?
### Use the Dropdown menu below to select a country.

### Answer:

```python
country_list = df[(df.country.isna() == False) & (df.country_count == 1)].country.sort_values().unique().tolist()

country_select = widgets.Dropdown(
                    options = country_list,
                    value = 'United States',
                    description = 'Countries',
                    disabled = False)

display(country_select)

button = widgets.Button(description="TV Shows")
button1 = widgets.Button(description="Movies")
output = widgets.Output()
output1 = widgets.Output()

display(button, output)
display(button1, output1)

def on_button_clicked(b):
    tv_df = df[(df.country.str.contains(country_select.value)) & (df.type == 'TV Show')].sort_values('averageRating', ascending=False)
    movie_df = df[(df.country.str.contains(country_select.value)) & (df.type == 'Movie')].sort_values('averageRating', ascending=False)
    with output:
        plt.title(label='Top 5 TV Shows in ' + country_select.value)
        sns.barplot(x=tv_df.averageRating.head(5), y=tv_df.title.head(5))
        #sns.barplot(movie_df.averageRating.head(5), movie_df.title.head(5))
        
def on_button_clicked1(b):
    tv_df = df[(df.country.str.contains(country_select.value)) & (df.type == 'TV Show')].sort_values('averageRating', ascending=False)
    movie_df = df[(df.country.str.contains(country_select.value)) & (df.type == 'Movie')].sort_values('averageRating', ascending=False)
    with output1:
        #sns.barplot(tv_df.averageRating.head(5), tv_df.title.head(5))
        plt.title(label='Top 5 Movies in ' + country_select.value)
        sns.barplot(x=movie_df.averageRating.head(5), y=movie_df.title.head(5))

button.on_click(on_button_clicked)
button1.on_click(on_button_clicked1)
```
